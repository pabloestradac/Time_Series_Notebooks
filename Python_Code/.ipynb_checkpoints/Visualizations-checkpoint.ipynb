```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
# Load dataset
series = pd.read_csv('data/daily-total-female-births.csv', header=0, parse_dates=[0], index_col=0, squeeze=True)
series.head()
```

```python
# Query a dataset using a date-time index
series['1959-01']
```

```python
# Dimension of a time series
series.size
```

```python
# Create a line plot
series.plot()
plt.show()
```

```python
# Create a dot plot
series.plot(style='k.')
plt.show()
```

```python
# Create a histogram plot
series.hist()
plt.show()
```

```python
# Create a density plot
series.plot(kind='kde')
plt.show()
```

```python
# create a boxplot of yearly data
groups = series.groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.boxplot()
plt.show()
```

```python
# create a boxplot of monthly data
groups = series.groupby(pd.Grouper(freq='M'))
months = pd.concat([pd.DataFrame(x[1].values) for x in groups], axis=1)
months = pd.DataFrame(months)
months.columns = range(1,13)
months.boxplot()
plt.show()
```
