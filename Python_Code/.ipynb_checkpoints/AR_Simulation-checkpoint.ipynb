```python
# Import libraries
import numpy as np
import numpy.random as npr
import numpy.linalg as la
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.optimize import minimize
```

```python
# Seed and warnings
npr.seed(1234)

import warnings
warnings.filterwarnings("ignore")

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

First, let's simulate an AR(1) process with mean 1, standard deviation 3 and autoregressive coefficient 0.8.
$$
y_t = \phi y_{t-1} + e_t
$$
where $\mu_y=1$, $\sigma_e=3$ and $\phi=0.8$.

```python
def AR(size=100, phi=0.8, mu_y=0, sig_e=1):
    var_y = sig_e**2 / (1-phi**2)
    e = sig_e * npr.normal(0, 1, (T+1, 1))
    y = np.zeros((T+1, 1))
    y[0] = np.sqrt(var_y) * npr.normal() + mu_y
    for t in range(1, T+1):
        y[t] = mu_y*(1-phi) + phi*y[t-1] + e[t]
    return y
```

```python
# Define parameters
T = 100
phi = 0.8
mu = 1
sigma = 3
# Create AR(1) process
y = AR(size=T, phi=phi, mu_y=mu, sig_e=sigma)
```

- Plot the simple and partial autocorrelation functions. 

```python
plot_acf(y)
plt.show()
```

```python
plot_pacf(y)
plt.show()
```

- Now, we're going to simulate the model 500 times for samples of size 200 and estimate the three parameters (constant, phi, and sigma_e) by conditional maximum likelihood for each simulation.

```python
def ar_loglik(param, T, y):
    c = param[0]
    phi = param[1]
    sigma2_e = param[2]
    tln2pi = (T-1) / 2.0 * np.log(2.0 * np.pi)
    tlnsig2 = (T-1) / 2.0 * np.log(sigma2_e)
    resid = y[1:] - c - phi*y[:T]
    y_ylag = 1 / (2*sigma2_e) * np.dot(resid.T, resid)
    lik = - tln2pi - tlnsig2 - y_ylag
    return -lik[0][0]
```

```python
c_ML = []
phi_ML = []
sd_ML = []
T = 200
for rep in range(500):
    y = AR(size=T, phi=phi, mu_y=mu, sig_e=sigma)
    res = minimize(ar_loglik, x0=[1.0, 1.0, 1.0], args=(T, y), method="L-BFGS-B")
    c = res.x[0]
    phi = res.x[1]
    sig2 = res.x[2]
    c_ML.append(c)
    phi_ML.append(phi)
    sd_ML.append(np.sqrt(sig2))
```

- Then, we're going to estimate the three parameters of this model (constant, phi, and sigma_e) by ordinary least squares for each simulation. 

```python
def ar_ols(y):
    T = y.shape[0] - 1 
    x = np.hstack((np.ones((T, 1)), y[:T]))
    xTx = np.dot(x.T, x)
    xTy = np.dot(x.T, y[1:])
    betas = np.dot(la.inv(xTx), xTy)
    e = y[1:] - np.dot(x, betas)
    sig2 = np.dot(e.T, e) / (T-2)
    return [betas[0][0], betas[1][0], sig2[0][0]]    
```

```python
c_OLS = []
phi_OLS = []
sd_OLS = []
T = 200
for rep in range(500):
    y = AR(size=T, phi=phi, mu_y=mu, sig_e=sigma)
    res = ar_ols(y)
    c = res[0]
    phi = res[1]
    sig2 = res[2]
    c_OLS.append(c)
    phi_OLS.append(phi)
    sd_OLS.append(np.sqrt(sig2))
```

- Finally, we will plot a histogram of the distribution of the estimates of the three parameters of the model, both for the Maximum Likelihood estimates and the OLS estimates, and compare its center to the true values of the parameters.

```python
phi_OLS = np.array(phi_OLS)
phi_ML = np.array(phi_ML)
c_ML = np.array(c_ML)
c_OLS = np.array(c_OLS)
sd_ML = np.array(sd_ML)
sd_OLS = np.array(sd_OLS)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.hist(phi_OLS, alpha=0.7, label="OLS")
ax1.axvline(phi_OLS.mean(), color='k', linestyle='dashed', linewidth=1, label="OLS")
ax1.hist(phi_ML, alpha=0.7, label="ML")
ax1.axvline(phi_ML.mean(), color='k', linewidth=1, label="ML")
ax1.axvline(0.8, color='r', linewidth=1.5, label="phi")
ax1.set_title("Histogram: phi")
ax1.legend(loc='upper right')

ax2.hist(c_OLS, alpha=0.7, label="OLS")
ax2.axvline(c_OLS.mean(), color='k', linestyle='dashed', linewidth=1, label="OLS")
ax2.hist(c_ML, alpha=0.7, label="ML")
ax2.axvline(c_ML.mean(), color='k', linewidth=1, label="ML")
ax2.axvline(1*(1-0.8), color='r', linewidth=1.5, label="phi")
ax2.set_title("Histogram: c")
ax2.legend(loc='upper right')

ax3.hist(sd_OLS, alpha=0.7, label="OLS")
ax3.axvline(sd_OLS.mean(), color='k', linestyle='dashed', linewidth=1, label="OLS")
ax3.hist(sd_ML, alpha=0.7, label="ML")
ax3.axvline(sd_ML.mean(), color='k', linewidth=1, label="ML")
ax3.axvline(np.sqrt(9 * (1-0.8**2)), color='r', linewidth=1.5, label="phi")
ax3.set_title("Histogram: sd")
ax3.legend(loc='upper right')

plt.show()
```
